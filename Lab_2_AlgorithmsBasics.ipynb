```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Sample function call: func1(5, 3)
import time
t1 = time.time()
x=print(func1(4,3)) # finding the multiplication of a and b i.e a*b
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
#the function performs multiplication using repeated addition
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a # finding the a to the power b i.e a**b
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
t1 = time.time()
print(func2(2,4))
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
   return result
```

```python
# Sample function call: func3(1234)
t1 = time.time()
print(func3(1234)) # finding the sum of digits in a number
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
t1 = time.time()
print(func4([1,2,3,4],[2,4,6,8]))
t2 = time.time() # counting the number of intersect in two arr
print("Time : ",(t2-t1) * 1000)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
t1 = time.time()
print(func5(16)) # determing weather it is perfect root of any number
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - i + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
t1 = time.time()
print(func8([3,1,2,7,8,6]))
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
t1 = time.time()
print(func12([1,2,3,4,5],2)) # left rotation of the array
t2 = time.time()
print("Time : ",(t2-t1) * 1000)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even = 0
  odd = 0
  for i in arr:
    if i%2==0:
      even += 1
    else:
      odd += 1
  return (even,odd)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,target):
  for i in arr:
    if i == target:
      return True
  return False
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
linear_Search([1, 3, 5, 7, 9],9)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,target):
  start = 0
  end = len(arr)-1
  while(start<end):
    mid = (start + end)//2
    if arr[mid] == target:
      return True
    elif target > arr[mid]:
      start = mid+1
    else:
      end = mid-1
  return False
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
Binary_Search([1, 3, 5, 7, 9,15,16,14,45],15)
```
